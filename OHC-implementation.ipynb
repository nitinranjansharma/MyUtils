```python
import pandas as pd
import numpy as np
def OHC_train(dataframe1,col_ids,train_test = 'Train'):
    
    
    dataframe = dataframe1.copy(deep = True)
    if train_test == "Train":
        #creation of label Encoding
        test_df = pd.DataFrame(list(set(list(dataframe[col_ids])))).reset_index()
        test_df.columns = ["Index","ID"]
        #global dict_values
        dict_values = (test_df.to_dict()['ID'])
        LabelEncoded = pd.merge(left = dataframe,right =test_df,how = 'left', left_on=col_ids, right_on = "ID").loc[:,['Index','ID']]        

        
        for i,j in dict_values.items():
            dataframe[j] = (LabelEncoded['Index'] == i) * 1
        #print(dict_values)    
        return(dataframe,(dict_values.items()))
```

```python
def OHC_test(dataframe1,col_ids,master_dict_vals,train_test = 'Test'):
    dataframe = dataframe1.copy(deep = True)
    if train_test == "Test":
        
        dict_values = master_dict_vals[col_ids]
        dict_values = dict(dict_values)
        try:
            test = dict_values
        except:
            print("Train the model first!")
            return()
        test_df = pd.DataFrame(list(set(list(dataframe[col_ids])))).reset_index()
        test_df.columns = ["Index","ID"]
        LabelEncoded = pd.merge(left = dataframe,right =test_df,how = 'left', left_on=col_ids, right_on = "ID").loc[:,['Index','ID']]        
        for i,j in dict_values.items():
            dataframe[j] = (LabelEncoded['Index'] == i) * 1
            
        return(dataframe)
```

```python

def OHC_master(df,train_test = "Train"):
    if train_test == "Train":
        global master_dict_vals
        master_dict_vals = {}
    cols = list(df.columns)
    
    for column in cols:
        if train_test == "Train":
            df, testd = OHC_train(df,column, train_test)
            master_dict_vals[column] = testd
            
            #print(master_dict_vals.items())
            
            
        if train_test == "Test":
            try:
                temp = master_dict_vals
                df = OHC_test(df,column, master_dict_vals,train_test)
            except:
                print("Train the model First")
                
            
    return(df)
            
   
    
```
